# LOGISTIC REGRESSION


What is Logistic Regression?

- Logistic regression is a supervised machine learning algorithm used for classification problems and it predicts the probability that input belongs to a particular class.


The name regression because it predicts the category class. Unlike in linear regression is used to predicts the number.

--------------------------------------------------------------------------------------------------------------------------------------------------------

**Example**

- Imagine we are a doctor. A patient comes for a checkup.

- We have the following information:

  Age = 55,

  Weight = 90 kg,
  
  Blood Pressure = 180,
  
  Sugar Level = High

- Now the doctor wants to answer one question:

  Does this patient have diabetes?

  The possible answer is

  - Yes or No.



Here we are predicting the category and not the numerical value.

--------------------------------------------------------------------------------------------------------------------------------------------------------

**FORMULA FOR LOGISTIC REGRESSION**


Z =  1 / (1 + e^-(b0+b1X))


Where,

b0- Intercept

b1- Slope

X - Independent variable


The output of the logistic regression will be the probablity value between 0 to 1.

For multiple independent variables, the logistic regression equation is

Z = 1 / 1+e−(b0+b1​X1​+b2​X2​+⋯+bn​Xn​)​

--------------------------------------------------------------------------------------------------------------------------------------------------------

*APPLICATIONS OF LOGISTIC REGRESSION*

Logistic regression is used whenever the output is categorical value.

1. Email- Spam or Not spam

2. In healthcare- Disease or No Disease

3. Bank- Loan approved or Loan Rejected

4. In Education - Pass or Fail

********************************************************************************************************************************************************

*ADVANTAGES OF LOGISTIC REGRESSION*

-  Easy to understand

-  Fast to train

-  Works well for binary classification

-  Gives probabilities (not just Yes/No)



********************************************************************************************************************************************************

*LIMITATIONS*


Logistic regression is having some limitations

-  Works best when the relationship is relatively simple.

-  Not suitable for very complex patterns.

-  Sensitive to outliers.

-  Performance decreases if features are highly correlated(multicollinearity).

--------------------------------------------------------------------------------------------------------------------------------------------------------

**REAL LIFE EXAMPLE**

Suppose a bank wants to predict whether a customer will repay a loan.

Based on Features:

- Salary
- Credit Score
- Age

The model first computes:

Score = -8 + (0.0001 × Salary)
          + (0.02 × Credit Score)
          - (0.03 × Age)

Assume the score is 2.

Then,

Logistic regression Z = 1 / 1+e−2​

The probability is about:0.88


So, 88% chance that the customer will repay the loan.

Since the obtained probability value is 0.88 which is greater than 0.5, so the customer will repay the loan.

**Prediction:** Loan Repaid
