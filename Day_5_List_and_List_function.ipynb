##Common List Functions

- `append`

Using 'append' function to add an element in a list

```python
fruits
```

```python
fruits.append('orange')
```

```python
fruits
```

The added element will always be appear at the end of the list

```python
# Adding two values in a list

fruits.append(['pineapple','mango'])
```

```python
fruits
```

Both the values will be appear as a list in a list

```python
# If we want to add multiple entries in a list, we can use 'for' loop to append.

item_1 = ['fruit1','fruit2','fruit3']

for i in item_1:
  fruits.append(i)
```

```python
fruits
```

- `insert`

`insert` function is used to insert a value in a specific position.

```python
fruits
```

```python
fruits.insert(2,'kiwi')   # insert function (position, value)
```

```python
fruits
```

```python
# Insert 'kiwi' in a list inside a list

fruits[5].insert(1,'kiwi')
```

```python
fruits
```
