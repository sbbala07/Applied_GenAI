SET AND SET OPERATIONS

What is set?
 - No duplicates - Automatically removes duplicates
 - Unordered - No defined order, or order may change

Creating a set

- Creating a set with curly brackets

```python
fruits = {'apple','banana','cherry'}

print(fruits)
```

Other way to create a set with specifically mention as set with a list of elements.

```python
numbers = set([1,2,3,4,5])

print(numbers)
```

**Removing item**

```python
fruits.remove('banana')
```

```python
fruits
```

Trying to remove an item which no longer exist.

```python
fruits.remove('grape')
```

It throws an key error since each element in a set stores as a key like dictionary. So it is trying to find key 'grape'.

*~ In order to avoid an error. we can use 'discard' function. It will trying to find key but it won't raises error.~*

```python
fruits.discard('apple')  # Using discard to remove
```

Never throws an error.

```python
fruits
```

**Set Operations**

```python
# Creating two sets

a = {1,2,3,4}
b = {3,4,5,6}
```

```python
a
```

`Union`

```python
a|b   # a union b
```

```python
a.union(b)  # Alternate method
```

It is trying to join unique elements in two sets.

`Intersection`

```python
a & b  # a intersection b
```

```python
a.intersection(b)  # Alternate method
```

It is takes the common values in both sets.

`Difference`

```python
a-b  # a difference b
```

```python
a.difference(b)  # alternate method
```

It means element in 'a' not in 'b'.

```python
b-a
```

It means element in 'b' not in 'a'.

`Symmetric Difference`

```python
a^b  # a symmetric difference b
```

```python
b^a  # b symmetric difference a
```

```python
a.symmetric_difference(b) # alternate method
```

```python
b.symmetric_difference(a)
```

It means elements in either a or b. Not in both. Here a^b and b^a meaning the same.
