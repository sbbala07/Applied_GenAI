- **Default function**

```python
print("Hello")  # Default function to print anything
```

- **Using functions to print "Hello"**

```python
def greet():  # Function definition
  print("Hello")
```

```python
greet()       # Function calling
```

- **Function with one parameter**

```python
def increment_by_10(a):  # Function definition
  print(a+10)
```

```python
increment_by_10(5)       # Function calling. Here is a is 5, then the result is 15
```

- **Functions with two parameters**

```python
def sum_(x,y):                          # Function definition. Here x and y are parameters
  print("Value of x:", x)
  print("Value of y:", y)
  print("Value of x+y:", x+y)
```

```python
sum_(2,3) # Default arguments. Here it takes x and y values as default
```

Here is takes the value of x as 2 and the value of y is 3.

```python
sum_(y=2,x=3) # Here we are specifically giving x and y values and it is called keyword arguments.
```

It takes the value of x and y as we specifically mentioned.

- **Function to return multiple values**

```python
def multi_values(a,b): # Function to print multiple values
  return a,b
```

```python
multi_values(3,4)
```

- Key Terms

```python
def fun_sample(x=2, y=4):  # Default parameters
  print("The value of x is:",x)
  print("The value of y is:",y)
```

```python
fun_sample()   # Here we already given the value of x and y as default
```

```python
fun_sample(10)   # Here it takes the value 10 as x and y as 4
```

```python
fun_sample(y=10)   # It takes x as 2 as default and y as 10 as mentioned here
```

```python
fun_sample(5,3)    # Here x is 5 and y is 3
```

- **Multi value function**

If we want to return the multiple parameters, we use mutiple keyword arguments

```python
def point_fun(*a):  # Here * indicates the muliple paramteters
  print(a)
```

```python
point_fun([10,5,12,24])  # Here we are passing a list with multiple values to print
```

```python
sum_(3,5)  # Adding the two values using sum_ function
```
