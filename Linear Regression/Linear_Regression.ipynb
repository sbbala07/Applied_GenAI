##LINEAR REGRESSION

What is Linear Regression?

- Linear Regression is a machine learning algorithm used to predict the continous numerical output by finding the relationship between the two variables.

- Linear Regression used to find the best fit line that explains the relationship between variables.

--------------------------------------------------------------------------------------------------------------------------------------------------------

The equation for linear regression is:

Y = mX + b

Here,

Y - Dependent variable (Output)

X - Independent variable

m - Slope (Howmuch Y changes when X increases)

b - Intercept ( Y value when X=0)

--------------------------------------------------------------------------------------------------------------------------------------------------------

**Example**

*Let us understand using the real life example*

We predict a student's exam score based on the number of hours they study


| Study Hours | Exam Score |
| ----------- | ---------: |
| 1           |         40 |
| 2           |         50 |
| 3           |         58 |
| 4           |         65 |
| 5           |         72 |


If we plot this data on a graph:

X-axis → Study Hours (input)

Y-axis → Exam Score (output)

Here, we notice study hours increases and exam score also increases. There is a linear relationship between the varibales.

Imagine if, Score=8(Study Hours)+32

Here, m- 8 and b - 32.

If the student studies for 6 hours.

then, Exam score = 8 x 6 + 32 = 80.

The predicted score is 80.

**Advantages**

- Easy to understand
- Easy to train
- Easy to interpret
- Less Computational cost


**Limitations**

- Assumption of linear relationship
- Sensitive to outliers
- Risk of underfitting
