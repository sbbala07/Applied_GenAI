Numpy Operations

Numpy array is a powerful, N-dimensional grid-like data structure used for high-performance numerical computing in Python where you can multiple values of same data type.

Advantages

- Use less memory than list when storing same type.
- Faster for numeric operations
- Keep values in order
- Easy index access

Disdvantages

- Same datatype only
- Less flexible than list
