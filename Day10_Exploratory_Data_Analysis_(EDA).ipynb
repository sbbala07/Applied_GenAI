**STEP : 8** - Outlier Detection

```python
# Outlier detection on Total bill column
```

- lower bound = q1 - 1.5 x IQR

- upper bound = q3 + 1.5 x IQR



Here, q1 - 25th qtr,

q3- 75th qtr

IQR- Interquartile range.


***Outlier - The datapoints which are falling below q1 and falling above than q3.***

```python
# Checking q1, q3 and IQR values for total column to findout the outliers

q1 = df['total_bill'].quantile(0.25)
q3 = df['total_bill'].quantile(0.75)
IQR = q3 - q1
```

Apply the values to findout lower bound and upper bound.

```python
lower_bound = q1 - (1.5*IQR)
upper_bound = q3 + (1.5*IQR)
```

```python
(df['total_bill'] < lower_bound) | (df['total_bill'] > upper_bound)
```

```python
df[(df['total_bill'] < lower_bound) | (df['total_bill'] > upper_bound)]   # below the lower bound or above the upper bound
```

There are 9 datapoints are falling below and above the bounds which are called outliers. Handling outliers are important which affect the model performances.
