DICTIONARY

- Dictionary is the key value pairs.
- It is a Fast look up data structure.

- Creating a Dictionary

```python
person = {
    'name' : 'alias',
    'age' : 25,
    'job': 'engineer'
}
```

It's a key value pairs. Here 'name','age','job' are keys. 'alias','25','engineer' are values.

```python
person
```

```python
empty = dict()   # Other way to create a dictionary.
```

```python
empty  # Empty dictionary without key and values
```

 Adding the element with key and value

```python
empty['name'] = 'alias'
```

```python
empty
```

Here 'name' is key and 'alias' is value.

```python
empty_1 = dict({
    'salary': 25000
})
```

```python
empty_1
```

- Accessing and updating the dictionary element

```python
person['name']  # Accessing an element with key
```

```python
person['age']=26   # Changing the value of element using key.
```

```python
person
```

Adding and removing keys

```python
person
```

```python
person['city']= 'chennai'  # Adding an element with key and values
```

```python
person  # Printing updated dictionary
```

```python
del person['job']   # Deleting an element with key
```

```python
person
```

Common Dictionary methods

```python
person.keys()   # Printing dictionary keys
```

```python
person.values()  # Printing dictionary values
```

```python
person.items()   # Printing both keys and values
```

- Looping through a dictionary

Using for loop we can loop through dictionary.

```python
for i in person.items():   # Go over dictionary both key and values
  print(i[0],i[1])         # Printing key and values
```

```python
for i in person.keys():   # Alternate method keys
  print(i,person[i])      # Printing key and values
```
