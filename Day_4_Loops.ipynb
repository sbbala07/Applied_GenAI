LOOPS

- Loops is a control statement used to execute the block of code repeatedly.

- It is used to reduce the lines of code with control flow statement.

Example

- PRINT VALUES FROM 1 TO 20 (Normal method)

```python
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
print(7)
print(8)
print(9)
print(10)
print(11)
print(12)
print(13)
print(14)
print(15)
print(16)
print(17)
print(18)
print(19)
print(20)
```

In order to avoid multiple lines of code. We use loop to execute the code.

WHILE LOOP / ENTRY CONTROLLED LOOP

- 'while' loop executes, until the condition is false.

```python
# Print 1 to 20

# Initialization
c = 1

# Condition
while c<=20:
  print(c)
  #Increment
  c = c+1
```

```python
# Print values from 20 1

c = 20 # Initialization

while c>=1: # Condition
  print(c)
  c = c-1   # Decrement
  print("After decrement", c)
```

FOR LOOP

- 'for' loop iterating over a sequence in list/tuple,set, range, dictionery

```python
# For loop with list

list = [1,"2",False,"Hello"]

for i in list:  # Iterating over each element in a list
  print(i)
```

```python
Fruits = ["apple",'banana','cherry']

for i in Fruits:  # Iterating over each element in a list
  print("I like",i)
```

```python
# For loop with range functions

for i in range(1,5):  # Using range functions to Print from 1 to 4.
  print(i)
```

```python
# Print 1 & 4

for i in range(1,5,3):  # (Start, stop, step)
  print(i)
```

```python
# Print from 1 to 5
for i in range(1,6):
  print(i)
```

```python
# Print only even numbers from 1 to 20

for i in range(1,21):
  if i%2==0:  # Getting only even no.
    print(i)
```

```python
# Print only even numbers without 'if' condition
for i in range(2,21,2): # Start, stop, step
  print(i)
```

```python
# Print multiples of 5 from 5 to 20.
for i in range(5,21,5):
  print(i)
```

```python
# Print 20 to 1 using for loop only range functions

for i in range(20,0,-1): # Reverse order from 20 to 1
  print(i)
```

LOOP CONTROL STATEMENT

It will change the normal execution of loop.

'CONTINUE' STATEMENT

- ''continue'' Skips the current iteration of the loop and moves to the next iteration

```python
# Print from 1 to 10 and skip 5.

for i in range(1,11):
  if i==5:
    continue # It will skip when i is 5.
  else:
    print(i)
```

'break' statement

- " break " terminates the loop entirely and come out of the loop.

```python
# print 1 to 11 but skip the loop before it reaches 5

for i in range(1,11):
  if i==5:
    break  # break the loop if i is 5.
  else:
    print(i)
```
