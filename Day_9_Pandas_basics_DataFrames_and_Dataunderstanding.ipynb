DATAFRAME OPERATIONS

- Creating DataFrame

Before we create a dictionary like data and convert that it into a DataFrame.

```python
data = {
    'name': ['alice','bob','charlie'],
    'age' : [25,30,35],
    'city': ['delhi','mumbai','chennai']
}
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

Here name, age and city are features and data are values.

- Saving the data as a csv file

Using to_csv to convert dataframe into a csv file.

```python
df.to_csv("data.csv", index = False)   # If index=False means giving only actual values instead of index
```

**Real-Life Example: Student Data**

students = pd.DataFrame({

    "ID": [101, 102, 103],

    "Name": ["Anu", "Ravi", "Meena"],

    "Marks": [88, 76, 90]

Creating a dictionary

```python
data_3 = {

"ID": [101, 102, 103],

"Name": ["Anu", "Ravi", "Meena"],

"Marks": [88, 76, 90]
}
```

Convert dictionary into a dataframe

```python
students = pd.DataFrame(data_3)
students
```

Or we can directly create a dataframe directly.

```python
students = pd.DataFrame({

"ID": [101, 102, 103],

"Name": ["Anu", "Ravi", "Meena"],

"Marks": [88, 76, 90]})
```

```python
students
```

```python
students.iloc[2]   # Accessing a row using index location
```

```python
students['Name']   # Accessing a column
```

```python
students.iloc[1:]    # Slicing the particular rows and columns
```

```python
students[['Marks','Name']]   # Printing two columns
```

```python
students.iloc[:,:]    # Alternate method to print entire dataframe. Where before the ',' is row and after ',' is column with start and stop.
```

```python
students.iloc[1:,:]   # Printing entire values from second row
```

```python
students.iloc[:,1:]   # Print entire values from second column
```

Records where marks > 80

```python
students
```

```python
students[students['Marks'] > 80]  # Apply some condition to filter out the data
```
