# AI, ML, DL, GENAI TERMINALOGIES

-  Let us understand each terminalogies with a practical day to day observations.

- For example, we own a restaurant and we hire few employees and each employee represents AI, ML, DL and GENAI.


................................................................................................................................................................................................................................................

**AI - The Smart Restaurant Manager**

The Role of manager is :

- Serves the customers quickly
- Reduce the customer waiting
- Increases sales
- Make customers happy

He make decisions and he doesnt need to know how to do, he just needs to act smart.

This is AI.

" AI is creating systems that capable of performing tasks that normally requires humans intelligence".

**ML- The waiter**

The Role of waiter:

If customer visits the restaurant, he doesnt know what customer wants at first.

After 100 days, the waiter notices.

- Students order burgers
- family orders pizza
- Business people orders coffee/Tea

So, the waiter learns from his experience. He learned from observing.

That is ML.

" Machine learning enables the systems learning patterns from data without explicitly programmed".

**DL- The experienced waiter**

The role of experienced waiter:

If the customer visits the restaurant, while looking customer face, mood, voice and expressions. He understands what customer wants.

Based on this experience, without customer asks, he instantly understand and proposing Coffee/Tea while looking at customers face.

This is Deep Learning(DL).

" Deep learning uses neural networks with multiple layers understand complex patterns from large amount of data"

**GENAI - The Chef who creates New dishes**

The role of chef here is:

He creates something new which never exists before.

For example, if a customer orders

*"I want a pizza mixed with burger."*

The chef creates new and that never exists before

That is GENAI.

" GENAI is the branch of AI that generates something new like images, audio, videos or texts".
