IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

DATA VISUALIZATION OPERATIONS

STEP: 1 - Loading the Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")  # Reading the csv file
df.head()
```

**COUNT THE NO.OF RECORDS FOR EACH DAY**

```python
df.groupby('day')['day'].count()
```

```python
df['day'].count()
```

```python
df.groupby('day')['day'].count().plot(kind = 'bar')  # Inbuilt python visulization technique

plt.xlabel('DAY')  # X axis label
plt.ylabel('COUNT') # Y axis label
plt.title("COUNT OF RECORDS BY DAY")  # Title of the chart
plt.show()  # Remove unnecessary texts
```

```python
# Using seaborn library to visualise the count of records by day

sns.countplot(x='day', data = df)
plt.title("COUNT OF RECORDS BY DAY")
plt.show()
```

*More visitors on saturday and least visitors on friday.*

Bar chart- Frequency or Numerical values with categorical columns.

```python
# Using line chart to see the total bill over time
from matplotlib.lines import lineStyles   # import matplotlib library
plt.plot(df['order_number'], df['total_bill'], marker = "*",color= 'red',markersize=12,linestyle ="dashed") # Plot line chart
plt.xlabel("ORDER NUMBER")
plt.ylabel("TOTAL BILL")
plt.title("TOTAL BILL OVER TIME")
plt.show()
```

```python
# Using plotly library to visualise total bill over time
import plotly.express as ps
ps.line(df, x = 'order_number', y = 'total_bill', markers = True)
```

**How does average total bill change across days?**

```python
df.head(2)
```

```python
# Finding the average bill

df.groupby('day')['total_bill'].mean()
```

```python
# Plot average bill on each sunday

avg_bill = df.groupby('day')['total_bill'].mean()
plt.plot(avg_bill.index, avg_bill.values, marker = "o",color= 'green',markersize=10,linestyle ="dotted")
plt.xlabel("Days")
plt.ylabel("Average bill")
plt.title("Average bill across days")
plt.show()
```

*Here, the average bill on sunday is higher. So customer spend more on sundays.*

**Which day has highest average tip?**

```python
# Average tip on each day
df.groupby('day')['tip'].mean()
```

```python
# Let visualise the highest average tip using bar chart using seaborn

sns.barplot(
    x = 'day',   # X axis label
    y = 'tip',   # Y axis label
    data = df    # Dataframe
)

plt.title("Average tip by day")  # Title
plt.show()
```

*Sunday has the highest average tip.*

**How are bills distributed?**

```python
# Visualise the distribution using histogram

fig = ps.histogram(
    df,
    x = 'total_bill', # X axis label
    nbins=10,  # No.of bars
    title = "Bill Distribution"
)

fig.show()
```

*Many customer spend between 10 to 20 range*

**Does higher bill mean higher tip?**

```python
# Let check the relation between the bill and tip using scatterplot

sns.scatterplot(
    data = df,
    x = 'total_bill',
    y = 'tip'
    )
plt.title("Bill Vs Tip")
plt.show()
```

*The points moving upward direction, so the total bill increases, the tip is also increasing. It has the positive correlation.*

**How does bill and tip vary with group size?**

```python
fig = ps.scatter(
    df,
    x = 'total_bill',
    y = 'tip',
    size ='size',
    color = 'day',
    title = 'Bubble chart'
)

fig.show()
```

*The bigger bubble means large group and higher bubble means higher tip.Here, the large group spent more paid more tips.*

**What percentage of customers are smokers?**

```python
# Plot pie chart to visualise the proportion of the smokers

smoker_count  = df['smoker'].value_counts()

plt.pie(
    smoker_count.values,
    labels = smoker_count.index,
    autopct='%1.1f%%'

)
plt.title("Smoker proportion")
plt.show()
```

*Here, 61.9% of customers are non smokers.*

**Which features are strongly related?**

```python
# Using heatmap to see the relations between the features

corr = df.corr(numeric_only = True)

sns.heatmap(
    corr,
    annot = True,
    cmap = 'coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

*Total bill and tip strongly related positvely.*
