```python
# Defining variables
x = 1
y = 10
```

IF STATEMENT

- Checking the condition using 'if' statement.

```python
if x>y:
  print("If statement executed!")
```

```python
x>y   # It gives boolean result
```

```python
if x<y:
  print("If statement executed!")
```

IF ELSE STATEMENT

- There are two conditions. 'if' executes if the condition is true, if not 'else' will executes.

```python
if x>y:
  print("If statement executed!")
else:
  print("Else statement executed!")
```

- Checking else alone will execute without 'if' condition.

```python
else:
  print("Else statement executed!")
```

Else alone won't execute.

IF-ELIF-ELSE-LADDER

- Checking multiple conditions. 'if' condition is fails, then 'elif' will execute. 'elif' also fails then 'else' statement will execute.

```python
# Building a grade calculation system

# "A" with Distinction, "A","B","C","D","E","F"
```

```python
mark = 0

if mark>=95 and mark<=100:
  print("A with distinction")
elif mark<95 and mark>=80:
  print("A grade")
elif mark>=70 and mark<80:
  print("B grade")
elif mark>=60 and mark<70:
  print("C grade")
elif mark>=50 and mark<60:
  print("D grade")
elif mark>=35 and mark<50:
  print("E grade")
elif mark<35 and mark>=0:
  print("F grade")
else:
  print("Invalid mark")
```

- Checking 'elif' statement alone will execute.

```python
elif mark>=35 and mark<50:
  print("E grade")
```

It gives error. 'elif' alone won't execute.

- Checking 'elif' and 'else' alone execute without 'if' statment.

```python
elif mark>=35 and mark<50:
  print("E grade")
else:
  print("F grade")
```

It won't execute without 'if' statement.

NESTED CONDITIONS

- Nested conditions means there are conditions inside the condition.

```python
mark = 0

if mark>=80 and mark<=100:
  if mark>=95:
    print("A with distinction")
  else:
    print("A grade")

elif mark>=70 and mark<80:
  print("B grade")
elif mark>=60 and mark<70:
  print("C grade")
elif mark>=50 and mark<60:
  print("D grade")
elif mark>=35 and mark<50:
  print("E grade")
elif mark<35 and mark>=0:
  print("F grade")
else:
  print("Enter valid mark")
```

REAL WORLD PROBLEM STATEMENT

- Real world problem statement using if-elif-else statement.

```python
temperature= 30

if temperature>35:
  print("Too hot")
elif temperature>=20:
  print("Nice")
else:
  print("Cold")
```
